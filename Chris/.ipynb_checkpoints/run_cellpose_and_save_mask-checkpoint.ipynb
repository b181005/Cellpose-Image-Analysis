# def run_cellpose_and_save_mask(ch2_path, save_dir):

````python
import os
import numpy as np
from cellpose import plot
from skimage.io import imread
from pathlib import Path
from cellpose import models
import matplotlib.pyplot as plt


model = models.CellposeModel(gpu=True)

def run_cellpose_and_save_mask(ch2_path, save_dir):
    """
    Performs Cellpose segmentation and saves only the resulting mask array (.npy)
    and the mask overlay image (.png).
    """
    print(f"\n--- Segmenting file: {os.path.basename(ch2_path)} ---")
    
    # 1. Load Images (as 2D grayscale)
    ch1_img = imread(ch2_path, as_gray=True)
    
    # 2. Run Cellpose (on the ch1/GFP channel)
    # The 'dat' variable is just used to capture the output tuple for convenience
    masks, flows, styles = model.eval(
        [ch1_img], 
        diameter=30, 
        flow_threshold=2, 
        cellprob_threshold=1
    )
    current_mask = masks[0]
    
    # Ensure the save directory exists
    save_path = Path(save_dir)
    save_path.mkdir(parents=True, exist_ok=True)
    
    file_stem = Path(ch2_path).stem
    
    # 3. Save the RAW MASK ARRAY (.npy file)
    # This is the most essential output for saving masks
    mask_npy_path = save_path / f"{file_stem}_mask.npy"
    np.save(mask_npy_path, current_mask)
    print(f"Mask array saved to: {mask_npy_path.name}")
    
    # 4. Save the MASK OVERLAY IMAGE (.png) for visual check
    mask_RGB = plot.mask_overlay(ch1_img, current_mask)
    overlay_path = save_path / f"{file_stem}_overlay.png"
    plt.imsave(overlay_path, mask_RGB)
    print(f"Mask overlay saved to: {overlay_path.name}")

    return current_mask, ch1_img # Return mask and image for further analysis
````

# example

````python
# Define a dedicated folder for the mask outputs
SAVE_FOLDER = 'C:/Users/labadmin/Documents/BZ-X800/Chris/Psychatg02_07112025/Cellpose_Masks'

# Run the function
mask_array, original_image = run_cellpose_and_save_mask(
    ch2_path='C:/Users/labadmin/Documents/BZ-X800/Chris/Psychatg02_07112025/B+_w1/W001/P00001/HM_W001_P00001_CH2.tif',
    save_dir=SAVE_FOLDER
)
````

````python
from skimage.measure import regionprops_table
import pandas as pd

# The mask_array and original_image are the outputs of your function
# mask_array, original_image = run_cellpose_and_save_mask(...)

props_table = regionprops_table(
    mask_array, 
    intensity_image=original_image, 
    properties=('label', 'area', 'mean_intensity') # These properties require both inputs
)

data_df = pd.DataFrame(props_table)

print(data_df.head())
````

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
# Use a categorical colormap (like 'nipy_spectral') to distinguish cell IDs
plt.imshow(mask_array, cmap='nipy_spectral') 
plt.title("Visualization of the Mask Array (Cell IDs)")
plt.colorbar(label='Cell ID')
plt.axis('off')
plt.show()
````

````python
from skimage.color import label2rgb
import matplotlib.pyplot as plt

# Create the colored overlay
overlay = label2rgb(
    mask_array, 
    image=original_image, 
    bg_label=0, 
    alpha=0.3 # Adjust alpha for transparency (0.3 is 30% opaque)
)

plt.figure(figsize=(10, 10))
plt.imshow(overlay)
plt.title("Filled Mask Overlay")
plt.axis('off')
plt.show()
````
