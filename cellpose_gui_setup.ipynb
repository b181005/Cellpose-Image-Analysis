## GPU Usage Tips

- **GPU Segmentation**: Use `segmentation_utils.py` functions which automatically detect and use GPU
- **Monitor GPU**: Open Task Manager (Ctrl+Shift+Esc) → Performance → GPU to monitor usage during processing
- **Adjust Model**: Use `diameter=30` for typical cells; reduce for smaller cells, increase for larger
- **Flow Threshold**: Adjust `flow_threshold` (default: 2) for different sensitivity
- **Batch Processing**: Use `batch_process_cellpose()` for multiple images

## Troubleshooting

- **GPU not detected**: Run the first cell and check output
- **Out of memory**: Reduce image size or batch size
- **Slow processing**: Verify `nvidia-smi` shows GPU memory usage
- **GUI won't launch**: Ensure notebook kernel is set to "Cellpose GPU"

```python
# Cell 4 — Example: Test GPU Segmentation (Optional)
import torch
import os
import numpy as np
from cellpose.models import CellposeModel
from PIL import Image

# Check if example images exist
example_dir = "../Example_Images"
if os.path.exists(example_dir):
    image_files = [f for f in os.listdir(example_dir) if f.endswith(('.tif', '.tiff', '.png', '.jpg'))]
    if image_files:
        # Load first image
        img_path = os.path.join(example_dir, image_files[0])
        img = Image.open(img_path)
        img_array = np.array(img)
        
        print(f"Testing GPU segmentation on: {image_files[0]}")
        print(f"Image shape: {img_array.shape}")
        print(f"GPU available: {torch.cuda.is_available()}")
        
        # Run Cellpose
        model = CellposeModel(gpu=torch.cuda.is_available())
        masks, flows, styles = model.eval(img_array, diameter=30)
        
        print(f"✓ Segmentation complete! Found {masks.max()} cells")
        print("GPU segmentation is working correctly!")
    else:
        print("No image files found in Example_Images folder")
else:
    print("Example_Images folder not found. Add test images to verify segmentation.")
```
