# Cellpose GUI Setup
Run **Cell 1** once to install, then **Cell 2** to launch the GUI.

```python
# Cell 1 — Install Cellpose with GUI support
import subprocess, sys

subprocess.run(
    [sys.executable, "-m", "pip", "install", "cellpose[gui]"],
    check=True
)
print("Done. Cellpose GUI installed.")
```

```python
# Cell 2 — Launch Cellpose GUI
import subprocess, sys

subprocess.Popen([sys.executable, "-m", "cellpose"])
print("Cellpose GUI launched — check your taskbar.")
```
